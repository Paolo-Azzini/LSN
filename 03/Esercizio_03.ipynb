### Prezzi delle opzioni europree di tipo **call**, e di tipo **put**, calcolati tramite simulazione montecarlo.

Come parametri sono stati usati: 
- asset price at $t=0$: $S(0)=100$
- delivery time: $T=1$
- strike price: $K=100$
- risk-free interest rate: $r=0.1$
- volatility: $\sigma=0.25$

I valori sono espressi come differenza dai valori esatti calcolati attraverso la soluzione analitica di Black-Scholes:
- call:  14.975790778311286
- put:  5.4595325819072364

```python
import matplotlib.pyplot as plt
import numpy as np
import math

prices = np.loadtxt("prices.dat")

plt.errorbar(prices[:,0]*1000,prices[:,1]-14.975790778311286,yerr=prices[:,2])
plt.xlabel('#step')
plt.ylabel('european call-options price')
plt.grid(True)
plt.show()

plt.errorbar(prices[:,0]*1000,prices[:,3]-5.4595325819072364,yerr=prices[:,4])
plt.xlabel('#step')
plt.ylabel('european put-options price')
plt.grid(True)
plt.show()
```
