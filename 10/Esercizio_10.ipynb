## Simulated annealing applicato alla risoluzione del problema del commesso viaggiatore

### Percorso circolare:

```python
import math
import numpy as np
import matplotlib.pyplot as plt

folder="circle_simulation_data/"

best = np.loadtxt("Simulated_Annealing/"+folder+"final_path.dat")

plt.figure(figsize=(10,10))
x, y=best[:,0],best[:,1]
plt.scatter(x, y)
for i in range(len(x)-1):
    plt.arrow(x[i], y[i], x[i+1]-x[i], y[i+1]-y[i])
plt.title("Best path")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()

lenght = np.loadtxt("Simulated_Annealing/"+folder+"lenght.dat")
lenght_gen = np.loadtxt("../09/"+folder+"Best_lenght.dat")

plt.figure(figsize=(10,10))
plt.plot(lenght, label="Simulated Annealing")
plt.plot(lenght_gen[:,1], label="Genetic Algorithm")
plt.xlabel("10^2 Matropolis steps / Generation")
plt.ylabel("Path lenght")
plt.legend(loc="best")
plt.grid(True)
plt.show()
```
