# Esercizio 04.1

### Equilibrazione:
Il sistema viene portato equilibrato attraverso re-esecuzioni successive usando la configurazione finale di una simulazione come configurazione successiva della simulazione successiva, per fare in modo che la temperatura del sistema coincida con la temperatura richiesta nell'input. Questa operazione viene ripetuta fino a che il valore dell'energia risulta stabile tra simulazioni successive.

<span style="color:green"> Nota: l'esecuzione successiva Ã¨ automatizzata tramite lo script "execution_script.h", il quale lancia la prima esecuzuione e 15 restart. </span>

```python
import matplotlib.pyplot as plt
import numpy as np
import math

ekin = np.loadtxt("MolecularDynamics_NVE/output_ekin.dat")
epot = np.loadtxt("MolecularDynamics_NVE/output_epot.dat")
etot = np.loadtxt("MolecularDynamics_NVE/output_etot.dat")
temp = np.loadtxt("MolecularDynamics_NVE/output_temp.dat")

fig=plt.figure(figsize=(16, 14))
plt.plot(ekin, label="K")
plt.plot(epot, label="V")
plt.plot(etot, label="E")
plt.plot(temp, label="T")
plt.legend(loc='best');
plt.grid(True)
plt.show()
```

# Esercizio 04.2, 3

Simulazione Argon ($\sigma = 0.34$ nm, $\epsilon/k_B = 120$ K, $m=39.948$ amu) 
## **Fase solida** ($\rho^\star = 1.1$, $T^\star = 0.8$ (cut-off radius: $r_c = 2.2$))

```python
T_conv=120
kb=1.380649E-23

folder="solid_data/"

ave_ekin = np.loadtxt(folder+"ave_ekin.dat")
ave_epot = np.loadtxt(folder+"ave_epot.dat")
ave_etot = np.loadtxt(folder+"ave_etot.dat")
ave_temp = np.loadtxt(folder+"ave_temp.dat")

plt.errorbar(ave_ekin[:,0],ave_ekin[:,2] * T_conv * kb,yerr=ave_ekin[:,3] * T_conv * kb, c='b')
plt.xlabel('#Block')
plt.ylabel('Kinetic Energy')
plt.grid(True)
plt.show()

plt.errorbar(ave_epot[:,0],ave_epot[:,2] * T_conv * kb,yerr=ave_epot[:,3] * T_conv * kb, c='y')
plt.xlabel('#Block')
plt.ylabel('Potential Energy')
plt.grid(True)
plt.show()

plt.errorbar(ave_etot[:,0],ave_etot[:,2] * T_conv * kb,yerr=ave_etot[:,3] * T_conv * kb, c='g')
plt.xlabel('#Block')
plt.ylabel('Total Energy')
plt.grid(True)
plt.show()

plt.errorbar(ave_temp[:,0],ave_temp[:,2] * T_conv, yerr=ave_temp[:,3] * T_conv, c='r')
plt.xlabel('#Block')
plt.ylabel('Temperature')
plt.grid(True)
plt.show()
```
