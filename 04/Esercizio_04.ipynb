```python
import matplotlib.pyplot as plt
import numpy as np
import math

ekin = np.loadtxt("MolecularDynamics_NVE/output_ekin.dat")
epot = np.loadtxt("MolecularDynamics_NVE/output_epot.dat")
etot = np.loadtxt("MolecularDynamics_NVE/output_etot.dat")
temp = np.loadtxt("MolecularDynamics_NVE/output_temp.dat")

fig=plt.figure(figsize=(16, 14))
plt.plot(ekin, label="K")
plt.plot(epot, label="V")
plt.plot(etot, label="E")
plt.plot(temp, label="T")
plt.legend(loc='best');
plt.grid(True)
plt.show()
```

```python
T_conv=120
kb=1.380649E-23


ave_ekin = np.loadtxt("MolecularDynamics_NVE/ave_ekin.dat")
ave_epot = np.loadtxt("MolecularDynamics_NVE/ave_epot.dat")
ave_etot = np.loadtxt("MolecularDynamics_NVE/ave_etot.dat")
ave_temp = np.loadtxt("MolecularDynamics_NVE/ave_temp.dat")

plt.errorbar(ave_ekin[:,0],ave_ekin[:,2] * T_conv * kb,yerr=ave_ekin[:,3] * T_conv * kb, c='b')
plt.xlabel('#Block')
plt.ylabel('Kinetic Energy')
plt.grid(True)
plt.show()

plt.errorbar(ave_epot[:,0],ave_epot[:,2] * T_conv * kb,yerr=ave_epot[:,3] * T_conv * kb, c='y')
plt.xlabel('#Block')
plt.ylabel('Potential Energy')
plt.grid(True)
plt.show()

plt.errorbar(ave_etot[:,0],ave_etot[:,2] * T_conv * kb,yerr=ave_etot[:,3] * T_conv * kb, c='g')
plt.xlabel('#Block')
plt.ylabel('Total Energy')
plt.grid(True)
plt.show()

plt.errorbar(ave_temp[:,0],ave_temp[:,2] * T_conv, yerr=ave_temp[:,3] * T_conv, c='r')
plt.xlabel('#Block')
plt.ylabel('Temperature')
plt.grid(True)
plt.show()
```
