## Algoritmo genetico per la risoluzione del problema del commesso viaggiatore

N.B.: Il codice è attualmente predisposto per la simulazione di 30000 generazioni,
simulazioni effettuate con un numero più elevato di generazioni non portano a un miglioramento del risultato mostrato.

### Città disposte su percorso circolare:

```python
import math
import numpy as np
import matplotlib.pyplot as plt

folder="circle_simulation_data/"

best = np.loadtxt(folder+"Best_path.dat")

plt.figure(figsize=(10,10))
x, y=best[:,0],best[:,1]
plt.scatter(x, y)
for i in range(len(x)-1):
    plt.arrow(x[i], y[i], x[i+1]-x[i], y[i+1]-y[i])
plt.title("Best path")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()

lenght = np.loadtxt(folder+"Best_lenght.dat")

plt.figure(figsize=(10,10))
plt.plot(lenght[:,0], label='Best individual')
plt.plot(lenght[:,1], label='Best half average')
plt.ylabel("Path lenght")
plt.xlabel("#Generations")
plt.legend(loc='best')
plt.grid(True)
plt.show()
```
