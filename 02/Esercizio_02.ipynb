# Esercizio 02.1


### l'integrale calcolato come media dei valori dell'integranda, campionati uniformemente nell'intervallo [0,1) .

Sono rappresentati i valori ottenuti in funzione del numero di blocchi considerati, ogni blocco consiste di $10^5$ step.

```python
import matplotlib.pyplot as plt
import numpy as np
import math

integral = np.loadtxt("integral.dat")

plt.errorbar(integral[:,0]*100000,integral[:,1],yerr=integral[:,2])
plt.xlabel('#step')
plt.ylabel('Integral')
plt.grid(True)
plt.show()

```

### Il risultato dello stesso integrale valutato campionando i valori con una distribuzione più "simile" all'integranda. 

Come distribuzione è stata scelta la retta $y=2(1-x)$. 

Uno sviluppo di Taylor dell'integranda sarebbe sicuramente una scelta più opportuna, tuttavia la cumulativa che si ottiene risulta difficilmente invertibile.

Nonostante la scelta non ottimale, il vantaggio della tecnica di importance sampling risulta evidente.

```python
plt.errorbar(integral[:,0]*100000,integral[:,3], yerr=integral[:,4])
plt.xlabel('#step')
plt.ylabel('Integral')
plt.grid(True)
plt.show()
```

Lo spostamento assoluto medio che aumenta come la radice nel numero di passi è indicativo di un comportamento diffusivo.

```python
import matplotlib.pyplot as plt
import numpy as np
import math

displacement = np.loadtxt("displacement.dat")

plt.errorbar(displacement[:,0],displacement[:,1],yerr=displacement[:,2])
plt.xlabel('#step')
plt.ylabel('Displacement')
plt.grid(True)
plt.show()
```
