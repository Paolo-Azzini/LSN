# Esercizio 05

Distribuzioni degli orbitali 1s e 2p dell'atomo di idrogeno, generate utilizzando l'algoritmo di Metropolis.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

p1s = np.loadtxt("1s_scatter.dat")
p2p = np.loadtxt("2p_scatter.dat")

fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(p1s[:,0],p1s[:,1],p1s[:,2], c=p1s[:,2], marker='.')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(10, 30)
plt.show()

fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(p2p[:,0],p2p[:,1],p2p[:,2], c=p2p[:,2], marker='.')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(10, 30)
plt.show()
```

```python
r1s = np.loadtxt("mean_radius_1s.dat")
r2p = np.loadtxt("mean_radius_2p.dat")

plt.errorbar(r1s[:,0]*1000,r1s[:,1],yerr=r1s[:,2])
plt.xlabel('#step')
plt.ylabel('<r/a0> in 1s orbital')
plt.grid(True)
plt.show()

plt.errorbar(r2p[:,0]*1000,r2p[:,1],yerr=r2p[:,2])
plt.xlabel('#step')
plt.ylabel('<r/a0> in 2p orbital')
plt.grid(True)
plt.show()
```
